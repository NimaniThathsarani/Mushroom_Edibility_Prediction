```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("secondary_data.csv" , sep=';')
print(df.columns.tolist())
```

```python
df.dtypes
```

```python
numeric_cols = ["cap-diameter", "stem-height", "stem-width"]
categorical_cols = [col for col in df.columns if col not in numeric_cols]
```

```python
# Convert data types
df[numeric_cols] = df[numeric_cols].astype(float)
df[categorical_cols] = df[categorical_cols].astype("category")

print("\nData Types After Conversion:\n", df.dtypes)
```

```python
df[numeric_cols].boxplot(figsize=(6,4))
plt.title("Boxplots of Numeric Features After Conversion")
plt.show()
```

```python
df[categorical_cols].nunique().plot(kind="bar", figsize=(10,4))
plt.title("Number of Unique Categories per Categorical Feature")
plt.show()
```
