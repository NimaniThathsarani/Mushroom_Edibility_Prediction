```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# Load dataset
df = pd.read_csv("secondary_data.csv", sep=';')
```

```python
numeric_cols = ["cap-diameter", "stem-height", "stem-width"]
```

```python
scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[numeric_cols] = scaler.fit_transform(df_scaled[numeric_cols])
```

```python
for col in numeric_cols:
    plt.figure(figsize=(10,4))
    plt.hist(df[col].dropna(), bins=30, alpha=0.5, label="Before")
    plt.hist(df_scaled[col].dropna(), bins=30, alpha=0.5, label="After")
    plt.title(f"{col} Distribution Before vs After Scaling")
    plt.legend()
    plt.show()

df_scaled[numeric_cols].boxplot(figsize=(6,4))
plt.title("Boxplots of Numeric Features After Scaling")
plt.show()
```
