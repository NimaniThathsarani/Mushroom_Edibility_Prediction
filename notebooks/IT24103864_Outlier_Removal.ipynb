```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("secondary_data.csv", sep=';')
```

```python
numeric_cols = ["cap-diameter", "stem-height", "stem-width"]
```

```python
df_outlier = df.copy()
for col in numeric_cols:
    Q1 = df_outlier[col].quantile(0.25)
    Q3 = df_outlier[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_outlier = df_outlier[(df_outlier[col] >= lower) & (df_outlier[col] <= upper)]
```

```python
for col in numeric_cols:
    plt.figure(figsize=(10,4))
    sns.boxplot(y=df[col])
    plt.title(f"{col} Before Outlier Removal")
    plt.show()

    plt.figure(figsize=(10,4))
    sns.boxplot(y=df_outlier[col])
    plt.title(f"{col} After Outlier Removal")
    plt.show()
```
